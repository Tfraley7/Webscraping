```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
url = 'https://stackoverflow.com/questions/tagged/python?sort=MostVotes&edited=true'
```

```python
# Assign path
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
browser.visit(url)
```

```python
html = browser.html
```

```python
#parse html
htmlsoup = soup(html, 'html.parser')
```

```python
#summary
htmlsoup.find('div', class_='s-post-summary--content-excerpt' )
```

```python
questiondivs = htmlsoup.find_all('div', class_='s-post-summary')
```

```python
#list of questions
print(questiondivs)
```

```python
questionlist = []
for question in questiondivs:
    summary = question.find('div', class_='s-post-summary--content-excerpt').text
    votes = question.find('span', class_='s-post-summary--stats-item-number').text
    questiondict = {}
    questiondict['summary'] = summary
    questiondict['votes'] = votes
    questionlist.append(questiondict)
questionlist
```
